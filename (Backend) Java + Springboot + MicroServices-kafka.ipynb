###### Default methods (Why we use it):
- Default methods enable you to add new functionality to existing interfaces without breaking existing implementations. (add methods with implementation by using default keyword).
-  In earlier versions of Java, adding a new method to an interface would force all classes that implement the interface to provide an implementation for the new method, which was not always feasible or desirable.
- This can help reduce the amount of repetitive code that needs to be written in classes that implement the interface
- In particular, default methods enable you to add methods that accept lambda expressions as parameters to existing interfaces.

###### Static methods in interface:
- Static methods in interfaces were introduced in Java 8 to allow interfaces to define and provide static methods that can be called without requiring any implementing class
- interface MyInterface {

    }
}
-  Default methods have an implementation but can be overridden by implementing classes. Static methods, on the other hand, cannot be overridden by implementing classes and are called on the interface itself.

#### Differnce between heap memory stack memmory
- __Heap memory__
 - heap memory is used for dynamic memory allocation, is managed explicitly by developers, and can have a longer lifetime. 
 - (eg. using new or malloc)
- __Stack memory__ :
 - Stack memory is used for static memory allocation, is managed automatically by the compiler, and is typically short-lived and associated with function calls.
 - Memory for local variables is automatically allocated when a function or method is called and automatically deallocated when it exits.
- In Java, you don't have explicit control over memory allocation and deallocation in either heap or stack memory; this is managed by the JVM. Local variables and method call information are typically stored in the stack, while objects and data structures are stored in the heap. The JVM takes care of garbage collection for objects in the heap, freeing up memory when objects are no longer reachable

#### Difference between final, finally and finalize in java
- __final__ :
 - final is a keyword used to declare a variable, method, or class as unchangeable or unextendable.
 - When used with a variable, it indicates that the variable's value cannot be changed after it has been assigned a value.
 - When used with a method, it signifies that the method cannot be overridden by subclasses.
 - When used with a class, it indicates that the class cannot be extended (subclassed).
- __finally__ :
 - finally is a block that is used with a try-catch block for exception handling.
 - This block is always executed, even if an exception is thrown or not.
- __finalize()__ :
 - finalize() is a method for object cleanup or resource release operations on an object just before garbage collection
 - The finalize method is automatically called by the garbage collector when it determines that there are no more references to the object.
 - use of finalize is discouraged in modern Java, and it's generally recommended to use other mechanisms, such as try-with-resources for resource management because it does not guaranteed for reliability and timing

#### Why Garbage Collection is necessary in Java?  What is the drawback of Garbage Collection?

- Garbage collection in Java helps programmers to focus on the development of the application instead of managing memory resources and worrying about memory leakage.
- Java Virtual Machine (JVM) automatically manages the memory periodically by running a garbage collector which frees up the unused memory in the application. Garbage collection makes Java memory efficient because it removes unreferenced objects from the heap memory.

#### What is a memory leak, and how does it affect garbage collection?

- In Java Memory leaks can be caused by a variety of factors, such as not closing resources properly, holding onto object references longer than necessary, or creating too many objects unnecessarily.
- There are situations in which garbage collector does not collect objects because there is a reference to those objects. In these situations where the application creates lots of objects and does not use them and every object has some valid references, a Garbage collector in Java cannot destroy the objects. These useless objects which do not provide any value to the program are known as Memory leaks
- Memory leaks can impact garbage collection negatively by preventing the garbage collector from reclaiming unused memory. This behavior will lead to slow performance or sometimes system failure. In a program, it is important to avoid memory leaks by managing resources and object references properly.

#### What are the Spring Boot Annotations?

- __@entity__ : In Java, the @Entity annotation is part of the Java Persistence API (JPA) specification, which is a standard for object-relational mapping (ORM). The @Entity annotation is used to mark a Java class as an entity, which means that instances of this class will be mapped to database tables.
When you use JPA and Hibernate (a popular JPA implementation), you can then create, read, update, and delete instances of the class (that is annoted with @entity), and these operations will be automatically translated into corresponding database operations. The @Entity annotation is crucial in defining the mapping between the Java class and the database table.
- __@SpringBootApplication__ : This is the main annotation used to bootstrap a Spring Boot application. It is typically placed on the main class of the application. It combines three annotations: @EnableAutoConfiguration, @Configuration and @ComponentScan. __Internally__, the @SpringBootApplication annotation is processed by the Spring Boot framework. The framework will first check if the annotation is present on the main class of your application. If it is, the framework will then enable auto-configuration, scan the packages for @Component annotated classes, and register the beans in the Spring container. __When__ you want to use auto-configuration, Typically used on the main class of a Spring Boot application, serving as the entry point, Includes @ComponentScan annotation to enable component scanning.
- __@EnableAutoConfiguration__ : When you want to customize auto-configuration, Can be used on any configuration class or in conjunction with @SpringBootApplication, Does not perform component scanning by itself. 
- __@Configuration__ : This annotation is used to indicate that a class contains configuration methods for the application context. It is typically used in combination with @Bean annotations to define beans and their dependencies.
- __@ComponentScan__ : One of the most important annotations in spring is @ComponentScan which is used along with the @Configuration annotation to specify the packages that we want to be scanned. @ComponentScan without arguments tells Spring to scan the current package and all of its sub-packages.
- __@Component__ : This annotation is the most generic annotation for any Spring-managed component. It is used to mark a class as a Spring bean that will be managed by the Spring container.
- __@RequestMapping__ : This annotation is used to map HTTP requests to a specific method in a controller. It can be applied at the class level to define a base URL for all methods in the class, or at the method level to specify a specific URL mapping. Handles various types of HTTP requests (GET, POST, etc.), @RequestMapping(value = /example, method = RequestMethod.GET)
- __@GetMapping__ : Specifically handles HTTP GET requests, @GetMapping(/example)
- __@ResponseBody__ :
- __@Controller__ : This is used to specify the controller
- __@RestController__ : This annotation is used to define a RESTful web service controller. It is a specialized version of the @Controller annotation that includes the @ResponseBody annotation by default. @RestController is a convenience annotation that combines @Controller and @ResponseBody. It indicates a controller where every method returns a domain object instead of a view.
- __@Autowired__ : This annotation is used to automatically inject dependencies into a Spring-managed bean. It can be applied to fields, constructor parameters, or setter methods.
- __@Service__ : This annotation is used to indicate that a class represents a service component in the application. It is typically used to annotate classes that contain business logic.
- __@Repository__ : This annotation is used to indicate that a class represents a data repository or data access object (DAO). It is typically used for classes that interact with the database or other persistent storage.


#### What are the scopes in Spring bean, Implement with code :
- In the Spring Framework, the scope of a bean defines how long a bean should exist and how it should be shared within the Spring container.

- __Singleton(Default)__ : This is the default scope in Spring. It means that the Spring container creates only one instance/object of the bean for each request and shares that single instance throughout the application. (For each object, there hashcode value will be the same)
- __Prototype__ : With this scope, the Spring container creates a new instance/object of the bean each time it is requested or injected. (For each object, there hashcode value will be different)
- __Request(Web)__ : This scope is specific to web applications. It creates a new bean instance for each HTTP request. The bean is destroyed once the request is processed.
- __Session(Web)__ : Similar to request scope, but it maintains a single instance throughout an HTTP session. The bean is created when a new session starts and is destroyed when the session ends.
- __Application(Web)__ : Also for web applications, this scope creates a single instance for the entire web application context. Its created upon application startup and destroyed when the application shuts down.
